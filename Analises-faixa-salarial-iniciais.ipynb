```python
# Dicionário de renomeação
novos_nomes = {
    'work_year': 'ano',
    'experience_level': 'nivel_experiencia',
    'employment_type': 'contrato',
    'job_title': 'cargo',
    'salary': 'salario',
    'salary_currency': 'moeda',
    'salary_in_usd': 'usd',
    'employee_residence': 'residencia',
    'remote_ratio': 'taxa_remoto',
    'company_location': 'empresa',
    'company_size': 'tamanho_empresa'
}

# Aplicando renomeação
df.rename(columns=novos_nomes, inplace=True)

# Verificando resultado
df.head()
```

```python
nivel_experiencia = {
    'SE': 'senior',
    'MI': 'pleno',
    'EN': 'junior',
    'EX': 'executivo'
}
df['nivel_experiencia'] = df['nivel_experiencia'].replace(nivel_experiencia)
df['nivel_experiencia'].value_counts()
```

```python
contrato = {
    'FT': 'integral',
    'PT': 'parcial',
    'CT': 'contrato',
    'FL': 'freelancer'
}
df['contrato'] = df['contrato'].replace(contrato)
df['contrato'].value_counts()
```

```python
tamanho_empresa = {
    'L': 'grande',
    'S': 'pequena',
    'M':	'media'

}
df['tamanho_empresa'] = df['tamanho_empresa'].replace(tamanho_empresa)
df['tamanho_empresa'].value_counts()
```

```python
taxa_remoto = {
    0: 'presencial',
    100: 'remoto',
    50: 'hibrido'
}

df['taxa_remoto'] = df['taxa_remoto'].replace(taxa_remoto)
df['taxa_remoto'].value_counts()
```

```python
import numpy as np

# importamos numpy para trabalhar com nulos
# novo dataframe criado para teste

df_salarios = pd.DataFrame({
  'nome' : ["Ana", "Bruno", "Carlos", "Daniele","Val"],
  'salario': [4000, np.nan, 5000, np.nan, 100_000]
})

# a linha seguinte substitui os nulos pela media arredondando

df_salarios['salario_media'] = df_salarios['salario'].fillna(df_salarios['salario'].mean().round(2))
# round é de arredondar com x casas

df_salarios['salario_mediana'] = df_salarios['salario'].fillna(df_salarios['salario'].median())

df_salarios
```

```python
df_temperaturas = pd.DataFrame({
    'dia': ['segunda', 'terça', 'quarta', 'quinta', 'sexta'],
    'temperatura': [30, np.nan, np.nan, 28, 27]
})

df_temperaturas['preenchido_ffill'] = df_temperaturas['temperatura'].ffill()

#ffill é de foward fill

df_temperaturas
```

```python
df_temperaturas = pd.DataFrame({
    'dia': ['segunda', 'terça', 'quarta', 'quinta', 'sexta'],
    'temperatura': [30, np.nan, np.nan, 28, 27]
})

df_temperaturas['preenchido_bfill'] = df_temperaturas['temperatura'].bfill()

#bfill é de back fill

df_temperaturas
```

```python
import plotly.express as px
```

```python
## Crie um gráfico de média salarial por senioridade em barras usando o plotly

senioridade_media_salario = df_limpo.groupby('nivel_experiencia')['usd'].mean().sort_values(ascending=False).reset_index()

fig = px.bar(senioridade_media_salario,
             x='nivel_experiencia',
             y='usd',
             title='Média Salarial por Senioridade',
             labels={'senioridade': 'Nível de Senioridade', 'usd': 'Média Salarial Anual (USD)'})

fig.show()
```

```python
remoto_contagem = df_limpo['taxa_remoto'].value_counts().reset_index()
remoto_contagem.columns = ['tipo_trabalho', 'quantidade']

fig = px.pie(remoto_contagem,
             values='quantidade',
             names='tipo_trabalho',
             title='proporção dos tipos de trabalho')
#grafico de pizza(torta)

fig.show()
```

```python
remoto_contagem = df_limpo['taxa_remoto'].value_counts().reset_index()
remoto_contagem.columns = ['tipo_trabalho', 'quantidade']

fig = px.pie(remoto_contagem,
             values='quantidade',
             names='tipo_trabalho',
             title='proporção dos tipos de trabalho',
             hole = 0.5)
#grafico de pizza(torta)
fig.update_traces(textinfo = 'percent+label')
fig.show()
```

```python
df_ds = df_limpo[df_limpo['cargo'] == 'Data Scientist']

ds_por_pais = df_ds['empresa'].value_counts().reset_index()
ds_por_pais.columns = ['pais', 'quantidade']

fig = px.bar(ds_por_pais,
             x='pais',
             y='quantidade',
             title='Quantidade de Cientistas de Dados por País')

fig.update_layout(xaxis_title='País',
                  yaxis_title='Quantidade')

fig.show()
```

```python
import pycountry

#função para converter ISO 2 para ISO3
def iso2_to_iso3(code):
  try:
    return pycountry.countries.get(alpha_2=code).alpha_3
  except:
    return None

#criar nova coluna com ISO 3
df_limpo['residencia_iso3'] = df_limpo['residencia'].apply(iso2_to_iso3)

#calcular media salarial por pais ISO3
df_ds = df_limpo[df_limpo['cargo'] == 'Data Scientist']
ds_por_pais = df_ds.groupby('residencia_iso3')['usd'].mean().reset_index()

#gerar o mapa
fig = px.choropleth(ds_por_pais,
                    locations = 'residencia_iso3',
                    color='usd',
                    color_continuous_scale= 'rdylgn',
                    title='Média Salarial de Cientista de Dados por País',
                    labels={'usd': 'Média Salarial Anual (USD)', 'residencia_iso3': 'País'})

fig.update_layout(geo=dict(showcoastlines=True,
                           projection_type="natural earth"
                    ))

fig.show()
```
